Function below:
- takes in a date string
- makes all sept sep
- tries to apply Mont. day, year format
- tries to apply day-month-year format
- returns new datetime object date_2

```python
def date_fun(date_1:str)->dt:
    date_1=date_1.replace('Sept', 'Sep')
    format_array = ['%b. %d, %Y', '%d-%b-%y', '%B %d, %Y']
    for f in format_array:
        try:
            return dt.strptime(date_1, f)
        except:
            continue

    raise Exception
```

```python
run_data["date_clean"]=run_data["date"].apply(date_fun)
run_data.head()
```
